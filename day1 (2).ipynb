1. Read a CSV file with data in multiple columns and compute mean, median mode of each of the columns.

```python
import pandas as pd
from statistics import median, mode


data_frame = pd.read_csv('/content/drive/MyDrive/students data.csv')

results = {}

for column in data_frame.columns:
    column_data = data_frame[column]
    column_mean = column_data.mean()
    column_median = median(column_data)
    column_mode = mode(column_data)

    results[column] = {
        'mean': column_mean,
        'median': column_median,
        'mode': column_mode
    }

for column, values in results.items():
    print(f"Column: {column}")
    print(f"Mean: {values['mean']}")
    print(f"Median: {values['median']}")
    print(f"Mode: {values['mode']}")
    print("---------------------------")
```

2. Plot each of the columns with different colors (check what kind of graph is suitable )

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/students data.csv')

colors = ['green', 'black', 'red']

for idx, column in enumerate(data.columns):
    plt.figure(figsize=(6, 4))
    data[column].plot(kind='bar', color=colors[idx])
    plt.xlabel(column)
    plt.ylabel('no of students')
    plt.title(f'BAR GRAPH for {column}')
    plt.show()
```

3. Create a game (like lottery/ rock paper/scissors) and write a python code to record how many times the computer won and the user won in an excel file. Plot the graph showing the success rate of user and computer with different colors

```python
import random
import pandas as pd
import matplotlib.pyplot as plt

gamelist = ['rock', 'paper', 'scissors']
comp_wins = 0
user_wins = 0
total_games = 0

results = []

while True:
    comp = random.choice(gamelist)
    user = input("Enter either rock or paper or scissors (or 'exit' to quit): ")

    if user == 'exit':
        break

    total_games += 1
    results.append(comp)

    if (comp == 'rock' and user == 'paper') or (comp == 'paper' and user == 'scissors') or (comp == 'scissors' and user == 'rock'):
        user_wins += 1
        results.append('user')
        print("You won")
    elif (comp == 'rock' and user == 'scissors') or (comp == 'paper' and user == 'rock') or (comp == 'scissors' and user == 'paper'):
        comp_wins += 1
        results.append('computer')
        print("Computer won")
    else:
        results.append('tie')
        print("TIE")

    print(f"Computer's choice: {comp}")
    print(f"Your choice: {user}")
    print(f"Computer wins: {comp_wins}")
    print(f"Your wins: {user_wins}")
    print(f"Total games played: {total_games}")

    df = pd.DataFrame({'Result': results})
    df.to_excel('game_results.xlsx', index=False)

    success_rates = [user_wins / total_games, comp_wins / total_games]

    plt.bar(['User', 'Computer'], success_rates, color=['green', 'black'])
    plt.xlabel('Players')
    plt.ylabel('Success Rate')
    plt.title('Success Rate of User vs. Computer')
    plt.show()
```
